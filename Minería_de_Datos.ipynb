```python
# Sirve para evaluar el desempño de una expresión en cuanto a tiempo
%timeit L = [n ** 2 for n in range(1000)]
```

```python
# Con doble porcentaje podemo evaluar el desempeño
# en cuanto tiempo de más de una línea de código
%%timeit
L = []
for n in range(1000):
    L.append(n ** 2)
```

```python
# Cuando el argumento de np.array es una iteración anidada
# genera una matriz
np.array([range(i, i + 3) for i in [2, 4, 6]])
```
